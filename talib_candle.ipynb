```python
import yfinance as yf
import pandas as pd

# Define the ticker symbol for NIFTY 50
nifty_ticker = "^NSEI"

# Download historical data for NIFTY 50
nifty_data = yf.download(nifty_ticker, start="2024-01-01", end="2024-08-14", interval="1d")

# Display the first few rows of the data
nifty_data.head()
```

```python
op = nifty_data['Open'].astype(float)
hi = nifty_data['High'].astype(float)
lo = nifty_data['Low'].astype(float)
cl = nifty_data['Close'].astype(float)
```

```python
nifty_data['DOJI'] = talib.CDLDOJI(op, hi, lo, cl)
nifty_data['Engulfing'] = talib.CDLENGULFING(op, hi, lo, cl)
nifty_data['HAMMER'] = talib.CDLHAMMER(op, hi, lo, cl)
nifty_data['MORNINGSTAR'] = talib.CDLMORNINGSTAR(op, hi, lo, cl)
nifty_data['MORNINGDOJISTAR'] = talib.CDLMORNINGDOJISTAR(op, hi, lo, cl) 
```

```python
import plotly.graph_objects as go
import pandas as pd

# Assuming nifty_data is your DataFrame containing the Open, High, Low, Close, and pattern data.

# Plot the candlestick chart
fig = go.Figure(data=[go.Candlestick(x=nifty_data.index,
                open=nifty_data['Open'],
                high=nifty_data['High'],
                low=nifty_data['Low'],
                close=nifty_data['Close'])])

# Set the vertical offset for the annotations
vertical_offset = 100  # Adjust this value for the vertical position

# Set the font size to ensure it fits within the candlestick width
annotation_font_size = 10  # Adjust this value to fit within the candle width

# Loop through the DataFrame to find patterns and add annotations
for i, row in nifty_data.iterrows():
    annotations = []
    
    # Check for each pattern and add to the annotations list
    if row['DOJI'] == 100:
        annotations.append("DOJI")
    if row['MORNINGSTAR'] == 100:
        annotations.append("Morning Star")
    if row['HAMMER'] == 100:
        annotations.append("Hammer")
    if row['MORNINGDOJISTAR'] == 100:
        annotations.append("Morning Doji Star")
    if row['Engulfing'] == 100:
        annotations.append("Engulfing")
    
    # If any annotations were found, place them vertically on the chart
    if annotations:
        fig.add_annotation(
            x=i,
            y=row['High'] + vertical_offset,  # Adjust y position to place above the candle
            text='<br>'.join(annotations),  # Join annotations with a line break to stack them
            showarrow=False,
            yshift=10,  # Adjust the vertical spacing
            textangle=90,  # Rotate text 90 degrees
            font=dict(size=annotation_font_size)  # Set the font size for the annotation
        )

# Show the plot
fig.show()
```
